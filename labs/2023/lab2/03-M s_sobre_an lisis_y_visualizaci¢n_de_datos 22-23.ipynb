## Conjuntos de datos incluidos en sklearn

El biblioteca scikit-learn inlcuye un módulo _datasets_ con varios conjuntos de datos "de juguete" que podemos utilizar para probar los algoritmos de aprendizaje incluidos en la biblioteca. 

En este ejemplo vamos a usar el famoso conjunto de datos IRIS que contiene ejemplos de plantas clasificadas en 3 familias (setosa, versicolour, virginica) según las longitudes de sus pétalos y sépalos: https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html

- __filename__: ruta del fichero local que contiene el conjunto de datos cargado
- __DESCR__: descripción del conjunto de datos en texto plano

## Creando un data frame de pandas

La biblioteca pandas permite trabajar con tablas de datos y proporciona herramientas básicas de análisis y visualización.

```python
import pandas as pd

df = pd.DataFrame(data=iris['data'], columns=iris['feature_names']) 
df
```

```python
# Añadimos una columna con la clase de cada individuo
df['species'] = iris['target']
dic_target = { 0 : 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(dic_target)

# Mostrar sólo las primeras filas de la tabla
df.head()
```

```python
# Tipos de datos de cada columna
df.dtypes
```

Ahora ya tenemos el conjunto de datos de sklearn cargado en un data frame y para analizar los datos, podemos proceder con los métodos vistos anteriormente.

```python
# La operación describe muestra datos estadísticos de cada columna
df.describe().transpose()
```

```python
# Podemos ver cómo cambian las variables para cada clase
df.groupby('species').describe()
```

```python
# Para facilitar la visualización podemos analizarlo para cada clase concreta, una a una, por ejemplo
df[df['species'] == 'setosa'].describe().transpose()
```

# Análisis bivariante 

El análisis bivariante, muestra la relación entre dos variables. 

Vamos a ver la correlación entre nuestras variables numéricas

```python
df.corr()
```

```python
# Ahora vamos a mostrar el diagrama de dispersión entre dos variables
df.plot.scatter(x='sepal length (cm)', y='petal width (cm)', c='Green')
```

```python
# Podemos usar el color como tercera dimensión para ver cómo se distribuyen los individuos de cada una de las clases
dic_colors = { 'setosa': 'blue', 'versicolor': 'red', 'virginica': 'green' }
colors = df['species'].map(dic_colors)

df.plot.scatter(x='sepal length (cm)', y='petal width (cm)', c=colors)
```

Si manejamos pocas variables podemos dibujar el diagrama de dispersión (scatter matrix) de todos los pares de variables para comprobar cómo están relacionadas visualmente.

```python
from pandas.plotting import scatter_matrix
%matplotlib inline

dic_colors = { 'setosa': 'blue', 'versicolor': 'red', 'virginica': 'green' }
colors = df['species'].map(dic_colors)

scatter_matrix(df, figsize = (12, 12), diagonal = 'hist', color=colors)
```

# Reescalado de las variables

En alguna ocasión las variables se encuentran en una escala muy distinta y esto puede distorsionar o perjudicar el funcionamiento de algunas técnicas. Para evitarlo se puede reescalar las variables.

Aunque el conjunto de datos del iris no sufre este problema, vamos a usarlo para practicar y observar el efecto resultante.

Probaremos con el reescalado usando los valores mínimo y máximo.

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

df[['sepal length (minmax)', 'petal width (minmax)']] = min_max_scaler.fit_transform(df[['sepal length (cm)', 'petal width (cm)']])

df.head()
```

Comprobamos el efecto en el gráfico de dispersión. Creamos el de referencia con las variables originales. 


```python
import matplotlib.pyplot as plt

scatter0 = df.plot.scatter(x='sepal length (cm)', y='petal width (cm)', c=colors)
```

```python


scatter1 = df.plot.scatter(x='sepal length (minmax)', y='petal width (minmax)', c=colors)

scatter1.axis('equal')


```

El cambio en la escala afectará a la distancia entre los puntos. 



# Escalado en dos pasos: escalando y volviendo a la escala original


Para ello usaremos el escalado por estandarización de los datos (restarles su media y dividirlos entre su desviación típica). Aunque se puede hacer con cualquiera de los métodos de escalado.

El escalado en dos pasos consiste en:
- Ajustar, que en el caso de la estandarización es calcular la media y desviación de los datos originales
- Transformar, que, dado el objeto `scaler`  ya ajustado, lo que hace es transforma el conjunto de datos que se le pase según el escalado correspondiente y su ajuste.



```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()

# Fase de ajuste
standard_scaler.fit(df[['sepal length (cm)']])

# Fase de transformación
df[['sepal length (standard)']] = standard_scaler.transform(df[['sepal length (cm)']])

df.head()
```

A continuación, podemos deshacer la transformación sobre los datos escalados, reestableciendo los valores originales. Para ello usamos `inverse_transform`

```python
df[['sepal length (redux)']] = standard_scaler.inverse_transform(df[['sepal length (standard)']])

df.head()
```

Aparentemente hemos hecho un ejercicio inútil. Sin embargo, cuando hagamos regresión, clasificación o clustering con las variables transformadas, será útil poder revertir la transformación para interpretar los datos resultantes.
